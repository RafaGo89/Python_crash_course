```python
# Creamos una tupla
dimensions = (200, 50)

# Accedemos a los valores de la tupla por sus Ã­ndices
print(dimensions[0])
print(dimensions[1])
```

```python
dimensions = (200, 50)

# Usamos un ciclo for para iterar los elementos de la tupla
for dimension in dimensions:
    print(dimension)
```

```python
dimensions = (200, 50)

print('Original dimensions:')

for dimension in dimensions:
    print(dimension)

# Aunque no se pueda cambiar una tupla, se pueden rehacer la tupla
dimensions = (400, 100)

print('\nModified dimensions:')

for dimension in dimensions:
    print(dimension)
```
