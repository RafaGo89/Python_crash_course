```python
current_number = 1
# Ciclo WHILE para contar los números del 1 al 5
while current_number <= 5:
    print(current_number)
    current_number += 1
```

```python
current_number = 0

while current_number < 10:
    current_number += 1
    # Si el número es par, se salta, no se imprime y se va al inicio del ciclo
    if current_number % 2 == 0:
        continue
    # Imprimimos solo los números impares
    print(current_number)
```
