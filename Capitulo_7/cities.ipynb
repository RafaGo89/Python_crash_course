```python
# Creamos un prompt de doble línea para decir al usuario que hacer
prompt = "\nEnter the name of a city you have visited:"
prompt += "\n(Enter 'quit' when you are finished) "

while True:
    city = input(prompt)

    if city == 'quit':
        # Si se ingresa 'quit' el programa termina
        break
    else:
        print(f"I'd love to go to {city.title()}")
```
