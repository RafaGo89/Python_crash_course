```python
pets = ['dog', 'cat', 'dog', 'goldfish', 'cat', 'rabbit', 'cat']
print(pets)

# Ejcutamos un bucle WHILE que se dentendr√° cuando eliminemos todas las
# veces que se repite 'cats'
while 'cat' in pets:
    # Eliminamos todas las apariciones de 'cats' en la lista
    pets.remove('cat')

print(pets)
```
