```python
# Escribimos primero una instrucción
prompt = 'If you share your name, we can personalize the message you see.'
# Concatenamos una segunda línea a esa instrución
prompt += '\nWhat is your first name? '

# Pasamos esas instrucciones como argumento a nuestra input()
name = input(prompt)
print(f'\nHello, {name}')
```
