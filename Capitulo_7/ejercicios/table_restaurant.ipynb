 Escriba un programa que pregunte al
usuario cuántos vienen a cenar. Si la respuesta es más de ocho, imprima un
mensaje diciendo al usuario que tendrán que esperar mesa. De lo contrario,
dígale que su mesa está lista.

```python
number_of_guests = input("How many are you, sir? ")
number_of_guests = int(number_of_guests)

if number_of_guests > 8:
    print("\nThen you'll have to wait a little bit, sir")
else:
    print("\nYour table is ready, sir")
```
