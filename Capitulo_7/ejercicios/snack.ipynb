Haga una lista llamada pedidos_bocadillos y rellénela con
los nombres de varios bocadillos. A continuación, haga una lista vacía
llamada bocadillos_terminados. Pase en bucle por la lista de pedidos de
bocadillos e imprima un mensaje para cada pedido, como "Su bocadillo de
atún está listo". A medida que se hagan los bocadillos, páselos a la lista de
terminados. Cuando todos los bocadillos estén hechos, imprima un mensaje
que los enumere todos.

```python
pedidos_bocadillos = ['bocadillo de atún',
                      'bocadillo de carne',
                      'bocadillo de frutas',
                      'bocadillo de camarón']
bocadillos_terminados = []

while pedidos_bocadillos:
    # Quitamos los bocadillos pedidos de la lista de pedidos
    bocadillo_actual = pedidos_bocadillos.pop()

    print(f'Su {bocadillo_actual} está listo')

    # Los añadimos a la lsita de bocadillos terminados
    bocadillos_terminados.append(bocadillo_actual)

print('\n--- Los bocadillos hechos fueron:')
for bocadillo in bocadillos_terminados:
    print(bocadillo)
```
