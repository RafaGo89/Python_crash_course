Escriba un programa que pregunte a los
usuarios por las vacaciones de sus sueños. Escriba unas instrucciones como
"Si pudieras visitar cualquier lugar del mundo, ¿dónde irías?". Incluya un
bloque de código que imprima el resultado de la encuesta.

```python
respuestas = {}

while True:
    # Pedimos el nombre y respuesta de la persona
    nombre = input('\n¿Cuál es tu nombre? ')
    respuesta = input('Si pudieras visitar cualquier lugar del mundo,'
    ' ¿a dónde irías? ')

    respuestas[nombre] = respuesta

    # Preguntamos si otra persona contestará la encuesta
    repetir = input('¿Otra persona contestará la encuesta (si/no) ')
    if repetir == 'no':
        break

# Mostramos los resultados de la encuesta
print('\n--- Resultados ---')
for nombre, respuesta in respuestas.items():
    print(f"A {nombre.title()} le gustaría ir de vacaciones a {respuesta.title()}")
```
