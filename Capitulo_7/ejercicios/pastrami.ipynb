Usando la lista pedidos_bocadillos del
ejercicio 7-8, asegúrese de que el bocadillo de 'pastrami' aparezca en la
lista al menos tres veces. Añada código al principio del programa para
imprimir un mensaje diciendo que no queda pastrami y use un bucle while
para eliminar todas las apariciones de 'pastrami' en pedidos_bocadillos.
Asegúrese de que no pasa ningún bocadillo de pastrami a la lista
bocadillos_terminados .

```python
pedidos_bocadillos = ['bocadillo de atún',
                      'bocadillo de pastrami',
                      'bocadillo de frutas',
                      'bocadillo de pastrami',
                      'bocadillo de camarón',
                      'bocadillo de pastrami']
bocadillos_terminados = []

print('No quedan bocadillos de pastrami :(\n')

while 'bocadillo de pastrami' in pedidos_bocadillos:
    pedidos_bocadillos.remove('bocadillo de pastrami')

while pedidos_bocadillos:
    # Quitamos los bocadillos pedidos de la lista de pedidos
    bocadillo_actual = pedidos_bocadillos.pop()

    print(f'Su {bocadillo_actual} está listo')

    # Los añadimos a la lsita de bocadillos terminados
    bocadillos_terminados.append(bocadillo_actual)

print('\n--- Los bocadillos hechos fueron:')
for bocadillo in bocadillos_terminados:
    print(bocadillo)
```
