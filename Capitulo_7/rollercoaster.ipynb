```python
height = input('How tall are you, in inches? ')
# Convertimos la entrada del usuario en número entero
height = int(height)

# Como la entrada es un número entero, podemos hacer comparaciones numéricas
if height >= 48:
    print('\nYou are tall enought to ride')
else:
    print("\nYou'll be able to ride when you're a little older")
```
