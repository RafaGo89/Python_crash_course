```python
# Creamos un prompt de doble línea para decir al usuario que hacer
prompt = "\nTell me something, and I will repeat it back to you:"
prompt += "\nEnter 'quit' to end the program "

# Variable para guardar el mensaje del usuario
message = ''
# Cuando ingresemos 'quit' el programa se detendrá
while message != 'quit':
    # Solicitamos el mensaje usando el prompt previamente definido
    message = input(prompt)
    if message != 'quit':
        print(message)
```

```python
# Creamos un prompt de doble línea para decir al usuario que hacer
prompt = "\nTell me something, and I will repeat it back to you:"
prompt += "\nEnter 'quit' to end the program "

# Creamos una bandera que mantendrá el programa activo mientras sea verdadera
active = True
while active:
    message = input(prompt)

    # Si se ingresa 'quit' la bandera se pone en False y se dentendrá en
    # la siguiente vuelta
    if message == 'quit':
        active = False
    else:
        # Si ingreso cualquier otra cosa, la imprimimos
        print(message)
```
