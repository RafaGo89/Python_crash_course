Cree un diccionario llamado ciudades. Use los nombres
de tres ciudades como claves en su diccionario. Cree un diccionario de
información sobre cada ciudad e incluya el país en el que se encuentra, su
población aproximada y alguna curiosidad sobre la ciudad. Las claves para cada ciudad serían país, población y curiosidad. Imprima el nombre de
cada ciudad y toda la información que tenga guardada sobre ella.

```python
cities = {'dublin':{
            'country': 'ireland',
            'population': 1458154,
            'fun fact': 'the Vikings founded Dublin in 841 AD'
            },
          'buenos aires':{
            'country': 'argentina',
            'population': 3121707,
            'fun_fact': 'tango was originally created there'
            },
          'glasgow':{
            'country': 'scotland',
            'population': 622820,
            'fun fact': "its subway is known as the 'Clockwork Orange'"
            }
          }

for city, facts in cities.items():
    print(f'{city.title()}')
    for key_fact, value_fact in facts.items():
        print(f'\t{key_fact.title()}: {value_fact}')
```
